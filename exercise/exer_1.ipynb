<span style = "color: Yellow">Exercise 5</span> . Suppose that after fitting a regression equation we obtain

Ŷ = 29.29 - 0.96X

The sample size is 8 and the standard error of the slope is 0.22. Use the 0.05 significance level. Can we conclude that the slope of the regression line is less than zero?

H₀: β₁ ≥ 0  (Slope is not negative or zero)

H₁: β₁ < 0   (Slope is negative) ← What we want to prove

```python
# t-statistic calculation
b1 = -0.96
SE = 0.22
t_statistic = b1 / SE

print(f"t-statistic = b₁ / SE = {b1} / {SE}")
print(f"t-statistic = {t_statistic:.4f}")
```

```python
#Critical t-value
t_critical = stats.t.ppf(0.05, df=6)  # One-tailed left test
print(f"Critical t-value (α=0.05, df=6, one-tailed left): {t_critical:.4f}")
```

Reject H₀ if: t-statistic < t-critical

-4.3636 < -1.9432 ✓ TRUE

```python
p_value = stats.t.cdf(t_statistic, df=6)  # One-tailed p-value
print(f"p-value = P(t ≤ {t_statistic:.4f}) = {p_value:.6f}")
```

```python
print("\nCONCLUSION:")
print("-" * 40)
print(f"1. t-statistic = {t_statistic:.4f}")
print(f"2. Critical value (α=0.05, one-tailed) = {t_critical:.4f}")
print(f"3. p-value = {p_value:.6f}")
print(f"4. Since t-statistic ({t_statistic:.4f}) < critical value ({t_critical:.4f})")
print(f"   and p-value ({p_value:.6f}) < α (0.05)")
print("\n REJECT H₀: I can conclude the slope is significantly less than zero.")
print("   The negative relationship is statistically significant.")
```

<span style = "color: Yellow">Exercise 6 </span>. Suppose that after fitting a regression equation we obtain

Ŷ = 1.85 - 0.08X

The sample size is 12, and the standard error of the slope is 0.03. Use the 0.05 significance level. Can we conclude that the slope of the regression line is different from zero?

H₀: β₁ = 0  (Slope is zero - no relationship)

H₁: β₁ ≠ 0   (Slope is different from zero) ← What we want to prove

```python
# t-statistic calculation
b1 = -0.08
SE = 0.03
t_statistic = b1 / SE

print(f"t-statistic = b₁ / SE = {b1} / {SE}")
print(f"t-statistic = {t_statistic:.4f}")
```

```python
# Critical t-values for two-tailed test
t_critical_left = stats.t.ppf(0.025, df=10)  # Left tail (2.5%)
t_critical_right = stats.t.ppf(0.975, df=10) # Right tail (97.5%)
print(f"Critical t-values (α=0.05, df=10, two-tailed):")
print(f"  Left tail: {t_critical_left:.4f}")
print(f"  Right tail: {t_critical_right:.4f}")
print(f"  Rejection region: t < {t_critical_left:.4f} or t > {t_critical_right:.4f}")
```

Reject H₀ if: |t-statistic| > |t-critical|

|-2.6667| > |2.2281| ✓ TRUE

2.6667 > 2.2281 ✓

```python
# p_value calculation
p_value = 2 * stats.t.cdf(t_statistic, df=10)  # Two-tailed: multiply by 2
# Since t is negative, take the left tail area × 2
print(f"p-value = 2 × P(t ≤ {t_statistic:.4f}) = {p_value:.6f}")
```

```python
print("\nfinal conclusion:")
print("-" * 40)
print(f"1. t-statistic = {t_statistic:.4f}")
print(f"2. Critical values (α=0.05, two-tailed) = ±{abs(t_critical_left):.4f}")
print(f"3. p-value = {p_value:.6f}")
print(f"4. Since |t-statistic| ({abs(t_statistic):.4f}) > critical value ({abs(t_critical_left):.4f})")
print(f"   and p-value ({p_value:.6f}) < α (0.05)")
print("\n REJECT H₀: I can conclude that the slope is significantly different from zero.")
print("   The relationship between X and Y is statistically significant.")
```

```python
# Use the original data arrays
age_data = np.array([8, 3, 6, 5, 5, 2, 8, 10, 9, 8, 2, 3, 2, 2])
cost_data = np.array([5555, 17888, 9963, 6793, 10774, 16310, 2475, 2808, 7073, 8978, 
                      11213, 9463, 15055, 20705])

# Test slope significance at α=0.10
slope, intercept, r, p, std_err = stats.linregress(age_data, cost_data)
t_stat = slope / std_err
df = len(age_data) - 2
t_crit = abs(stats.t.ppf(0.05, df))  # α=0.10 two-tailed
sig = abs(t_stat) > t_crit

print("(d) Slope Significance Test (α = 0.10):")
print(f"t-statistic = {t_stat:.4f}")
print(f"Critical t-value = ±{t_crit:.4f}")
print(f"p-value = {p:.4f}")
print(f"\nSignificant? {'YES' if sig else 'NO'}")
print(f"Interpretation: The slope is {'statistically' if sig else 'not statistically'} significant.")
```

**(a)** Determine the correlation between price and demand. Plot the data in a scatter diagram. Does the data seem to be linear?

**(b)** Transform the price to a log to the base 10. Plot the log of the price and the demand. Determine the correlation coefficient. Does this seem to improve the relationship between the variables?

<details>
  <summary><font size="3" color="gold"><b>(a) Determine the correlation between price and demand. Plot the data in a scatter diagram. Does the data seem to be linear?</b></font></summary>


```python
# Data
demand = np.array([2, 5, 8, 12, 16, 21, 27, 35, 45, 60])  # X variable
price = np.array([120, 90, 80, 70, 50, 45, 31, 30, 25, 21])  # Y variable
```

```python
# Calculate correlation
r, p_value = stats.pearsonr(demand, price)
print(f"Correlation coefficient (r) = {r:.4f}")
print(f"R-squared (r²) = {r**2:.4f} = {r**2*100:.1f}%")
print(f"Interpretation: {abs(r):.1%} {'negative' if r < 0 else 'positive'} correlation")
```

```python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(demand, price, color='blue', s=100, edgecolor='black', linewidth=1.5)
plt.xlabel('Demand (X)', fontsize=12)
plt.ylabel('Price (Y)', fontsize=12)
plt.title('Scatter Plot: Price vs Demand', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# linear regression
slope, intercept, r_line, p, std_err = stats.linregress(demand, price)
demand_line = np.linspace(min(demand), max(demand), 100)
price_line = intercept + slope * demand_line
plt.plot(demand_line, price_line, 'r--', alpha=0.7, linewidth=2, label=f'Linear fit: r={r:.3f}')
# Add equation text
equation = f'Price = {intercept:.1f} + {slope:.2f}×Demand'
plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes,
         fontsize=11, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.legend()
plt.tight_layout()
plt.show()
```

Yes, the data appears mostly linear because of several reasons below:
1. Strong negative correlation (r = -0.967)
2. Points follow a clear downward trend
3. 93.5% of price variation explained by demand
4. Linear regression line fits well through points

However, there might be slight curvature as demand increases,
suggesting a nonlinear relationship could fit even better.

<details>
  <summary><font size="3" color="gold"><b>(b) Which transformation from part (a) does the best job of producing an approximate linear relationship? Use the selected transformation to predict lead content when distance is 45m. </b></font></summary>

```python
# Transform
ln_x = np.log(x)
ln_y = np.log(y)

# Linear regression (ln_y = b0 + b1 * ln_x)
n = len(ln_x)
b1 = (n*np.sum(ln_x*ln_y) - np.sum(ln_x)*np.sum(ln_y)) / (n*np.sum(ln_x**2) - np.sum(ln_x)**2)
b0 = np.mean(ln_y) - b1*np.mean(ln_x)

print(f"Regression equation: ln(y) = {b0:.3f} + {b1:.3f} * ln(x)")
print(f"Or: y = {np.exp(b0):.2f} * x^{b1:.3f}")
```

```python
# Predict for x = 45
x_new = 45
ln_x_new = np.log(x_new)
ln_y_pred = b0 + b1 * ln_x_new
y_pred = np.exp(ln_y_pred)

print(f"\nFor x = {x_new}m:")
print(f"ln(y_pred) = {ln_y_pred:.3f}")
print(f"y_pred = {y_pred:.2f} ppm")
```

**Best transformation:** `ln(y) vs ln(x)` (log-log transformation)

**The Reason is:** This transformation produces the most linear scatter plot, indicating a power law relationship between distance and lead content.

This power law model makes physical sense: lead contamination decreases with distance following a decay pattern, where the rate of decrease slows down as you move farther from the highway.
