```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

<span style="color: gold">Data</span>

```python
Data = pd.read_excel('lab1.xlsx', sheet_name='Ex1') 
```

```python
Data
```

```python
x_values = Data.iloc[:, 0]
y_values = Data.iloc[:, 1]
```

```python
p_x = np.array(x_values, dtype=float)
f_y = np.array(y_values, dtype=float)
```

```python
#A scatter plot of the data
plt.scatter(p_x, f_y)
plt.xlabel('# people followed(followees)')
plt.ylabel('# followers')
plt.title('Scatter plot of followers vs people followed(followees)')  
plt.show()
```

<span style="color: gold"> Insight obtained</span> based on  the above plot, it shows a positive trend because the observations/slope moves upward. 

As followees tend to increase the followers increases. 

<details>
  <summary><font size="3" color="gold"><b>To Calculate correlation coefficient</b></font></summary>

```python
r = np.corrcoef(p_x, f_y)[0, 1]
print(f"Correlation coefficient: r = {r:7f}")
```

<span style = "color: gold"> Interpretation:</span>
based on this value of this correlation coefficient. it is a positive relationship. following more people are associated with more followers.

<details>
  <summary><font size="3" color="gold"><b>To Calculate regression equation</b></font></summary>

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(p_x, f_y)

print(f"   ŷ = {intercept:.2f} + {slope:.4f}x")
print(f"   or to write like this: Followers = {intercept:.2f} + {slope:.4f} × (People Followed)")
```

<span style = "color: gold">Interpretation:</span> 
**Intercept = 60.35:** That means if an AIMS student follows zero people(x =0), they will have about 60 - 61 folowers on an average. 

**Slope = 0.8937** means that for each additional person an AIMS student follows, their followers increases by 0.8937 on an average

<details>
  <summary><font size="3" color="gold"><b>To Forecast Instagram followers of an AIMS student who follows 200
people.</b></font></summary>

```python

y_pred = intercept + slope * 200
print(f"Prediction for 200: {y_pred:.4f}")
```

<span style = "color: gold"> Interpretation</span> for x = 200 

ŷ = 60.35 + 0.8937 × 200 ≈ 239

meaning that: A student who follows 200 people is to have 239 followers. 

<span style="color: gold">Insight derived:</span> The model shows that while following more people is associated with having more followers, the relationship is only moderately strong.

```python
Data2 = pd.read_excel('lab1.xlsx', sheet_name='Ex2')
```

```python
Data2
```

```python
x_values2 = Data2.iloc[:, 0]
y_values2 = Data2.iloc[:, 1]
```

```python
#Independent variable (# NbOfContacts)
#Dependent variable (# Sales)
x = np.array(x_values2, dtype=float)
y = np.array(y_values2, dtype=float)
```

```python
# A scatter plot of the data
plt.scatter(x, y)
plt.xlabel('# NbOfContacts')   
plt.ylabel('# Sales')
plt.title('Scatter plot of Sales vs Number of Contacts')
plt.show()
```

<details>
  <summary><font size="3" color="gold"><b> (a) Determine the regression equation. </b></font></summary>

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
print(f"   Estimated model: ŷ = {intercept:.3f} + {slope:.3f}x")
print(f"   or: Sales = {intercept:.3f} + {slope:.3f} × Contacts")

print(f"   The estimated intercept: β₀ = {intercept:.3f}")
print(f"   The estimated slope: β₁ = {slope:.3f}") 
```

<spam style ="color: gold">Interpretation:</span> 

This means that when the **Intercept** is -12.201 is, when contact = 0 the predcited sales are - 12. 2 but this is not realistic in practice.

**slope is 2.195** means tht for each additional contact, the sales inceases by about 2.2units. 

```python
plt.scatter(x, y)
plt.xlabel('# NbOfContacts')
plt.ylabel('# Sales')
plt.title('Scatter plot of Sales vs Number of Contacts')

x_line = np.array([min(x), max(x)])
y_line = -12.201 + 2.195 * x_line
plt.plot(x_line, y_line, color='red', label='Regression Line') 
plt.show()
```

<details>
  <summary><font size="3" color="gold"><b> (b)Estimated sales for 40 contacts </b></font></summary>

```python
contacts = 40
predicted_sales = intercept + slope * contacts
```

```python
print(f" For {contacts} contacts: The predicted sales = {predicted_sales:.3f}")
```

<span style = "color: gold" >This means that: </span> If 40 contacts are made,  the estimated sales are 75.6(Thousand dollar)
