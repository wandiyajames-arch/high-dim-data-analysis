```python
#Independent variable (# NbOfContacts)
#Dependent variable (# Sales)
x = np.array([14, 12, 20, 16, 46, 23, 48, 50, 55, 50])
y = np.array([24, 14, 28, 30, 80, 30, 90, 85, 120, 110])
```

```python
print(f"   Sales = {intercept:.3f} + {slope:.3f} × Contacts")
print(f"   or: ŷ = {intercept:.3f} + {slope:.3f}x")
```

<details>
