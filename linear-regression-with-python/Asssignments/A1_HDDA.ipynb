```python
import numpy as np
from scipy import stats
import pandas as pd
```

<span style="color: gold">Data</span>

```python
Data = pd.read_excel('lab1.xlsx', sheet_name='Ex1') 
```

```python
Data
```

```python
x_values = Data.iloc[:, 0]
y_values = Data.iloc[:, 1]
```

```python
p_x = np.array(x_values, dtype=float)
f_y = np.array(y_values, dtype=float)
```

<details>
  <summary><font size="3" color="gold"><b>To Calculate correlation coefficient</b></font></summary>

```python
r = np.corrcoef(p_x, f_y)[0, 1]
print(f"Correlation coefficient: r = {r:.4f}")
```

<details>
  <summary><font size="3" color="gold"><b>To Calculate regression equation</b></font></summary>

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(p_x, f_y)

print(f"   ŷ = {intercept:.2f} + {slope:.4f}x")
print(f"   or to write like this: Followers = {intercept:.2f} + {slope:.4f} × (People Followed)")
```

<details>
  <summary><font size="3" color="gold"><b>To Forecast Instagram followers of an AIMS student who follows 200
people.</b></font></summary>

```python

y_pred = intercept + slope * 200
print(f"Prediction for 200: {y_pred:.4f}")
```

<span style="color: gold">Insight derived:</span> The model shows that while following more people is associated with having more followers, the relationship is only moderately strong.

```python
Data2 = pd.read_excel('lab1.xlsx', sheet_name='Ex2')
```

```python
Data2
```

```python
x_values2 = Data2.iloc[:, 0]
y_values2 = Data2.iloc[:, 1]
```

```python
#Independent variable (# NbOfContacts)
#Dependent variable (# Sales)
x = np.array(x_values2, dtype=float)
y = np.array(y_values2, dtype=float)
```

<details>
  <summary><font size="3" color="gold"><b> (a) Determine the regression equation. </b></font></summary>

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
print(f"   Estimated model: ŷ = {intercept:.3f} + {slope:.3f}x")
print(f"   or: Sales = {intercept:.3f} + {slope:.3f} × Contacts")

print(f"   The estimated intercept: β₀ = {intercept:.3f}")
print(f"   The estimated slope: β₁ = {slope:.3f}")
```

<details>
  <summary><font size="3" color="gold"><b> (b)Estimated sales for 40 contacts </b></font></summary>

```python
contacts = 40
predicted_sales = intercept + slope * contacts
```

```python
print(f" For {contacts} contacts: The predicted sales = {predicted_sales:.3f}")
```

<span style = "color: gold" >This means that: </span> If 40 contacts are made, estimated sales are 75.6(Thousand dollar)
